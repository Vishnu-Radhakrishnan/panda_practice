```python
import  seaborn as sns
```

```python
t=sns.load_dataset('tips')
```

```python
t
t.columns
```

```python
#strip_plot
sns.stripplot(y='tip',data=t,jitter=True)
plt.show()
```

```python
#swarm_plot
sns.swarmplot(data=t,x='day',y='tip',hue='sex')
plt.show()

```

```python
#Joint_plot
sns.jointplot(data=t,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
#Density_plot
sns.jointplot(data=t,x='total_bill',y='tip',kind='kde')
plt.show()
```

```python
#pair_plot
sns.pairplot(t,hue='sex')
plt.show()
```

```python
train=pd.read_csv('/content/train.csv')
```

```python
train
```

```python
train.groupby('Survived')[['PassengerId']].count()
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.boxplot(data=train,x='Survived',y='Age')
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.barplot(data=train,x='Survived',y='Sex')
```

```python
sns.barplot(data=train,x='Pclass',y='Survived',hue='Sex')
```

```python
import matplotlib.pyplot as plt
```

```python
s1=train.loc[train['Survived']==1,'Age']
sns.distplot(s1)
plt.title='Survived'
```

```python
s2=train.loc[train['Survived']==0,'Age'].dropna()
sns.distplot(s2)
```

```python
grid=sns.FacetGrid(train,col='Survived',row='Pclass',height=2.5,aspect=1.7)
grid.map(plt.hist,'Age',alpha=0.9,bins=20)
grid.add_legend()
```
